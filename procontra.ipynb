```python
def get_all_petitions(url):
    links=[]
    t=get_comment_linkpages(url)
    if len(t)==1:
        req1 = get(t[0])
        html_soup1 = BeautifulSoup(req1.text, 'html.parser')
        for l in html_soup1.find_all('li', class_ = 'clearfix'):
            if l.h3 != None:
                #print(l.h3.find('a',class_ ='theme-link-variant-3'))
                x=l.h3.find('a',class_ ='theme-link-variant-3')
                links.append(x['href'])
    else:
        for l in t:
    
            req1 = get(l)
            html_soup1 = BeautifulSoup(req1.text, 'html.parser')
            for l in html_soup1.find_all('li', class_ = 'clearfix'):
                if l.h3 != None:
                    #print(l.h3.find('a',class_ ='theme-link-variant-3'))
                    x=l.h3.find('a',class_ ='theme-link-variant-3')
                    links.append(x['href'])
    return links
```

```python
def get_comment_linkpages(url_comment):
   # print(html_soup)
    
    req=get(url_comment)
    html_soup= BeautifulSoup(req.text, 'html.parser')
    c=html_soup.find("div",{'class':'module-generic-pagenav-numbers'})
    comment_page_links=[]
    if c is not None:
        
        #print("c",c)
        page_numbers=[]
        count=0
        #print(len(c))
        for a in c.find_all("a",href=True):
            #print ("Found the URL:", a['href'])
            #print ("/petition/kommentare/keine-abschiebungen-nach-syrien?sortierung=gewichtung&seite={}".format(x))
            page_numbers.append(a.text)

         #   print(a.text)
            count=count+1
        #print("Number of pages",page_numbers[-1])
        
        for x in range(int(page_numbers[0]),int(page_numbers[-1])+1):
            #print (url_comment+"&seite={}".format(x))
            comment_page_links.append(url_comment+"&seite={}".format(x))
    else:
        #print("url comment")
        comment_page_links.append(url_comment)
      #  return comment_page_links
        
    return comment_page_links  
```

```python
def link_metadata(url,i,file):
    req=get(url, stream=True)
    html_soup_abschiebungen_syrien=BeautifulSoup(req.text, 'lxml')
    pet_1=html_soup_abschiebungen_syrien.find("ul",{'id':'petition-submenu'})
    #pet_2=html_soup_abschiebungen_syrien.find("div",{'class':'petitionen-liste'})
    #print(pet_2.h2)
    l=[]
    for i in pet_1:
        x=i.find("a")
        if x != -1:
            x=x.text.strip()
            l.append(x)
            #print(x)
    #print("Anzahl der Pro und Contra Argumente: ",l[1].split()[-1])
    count=l[1].split()[-1]
    print("count",(count.isnumeric()))
    with open(file,'a') as f:
        if count.isnumeric()==False:
            
            count="Kein Argumente"
            f.write("Anzahl der Pro und Contra Argumente: "+count+'\n')
            f.write('\n'+'========================================'+'\n'+'\n')
        else: 
            #co=count
            f.write("Anzahl der Pro und Contra Argumente: "+count+'\n'+'\n')
            
    return count
            
        #f.write("Anzahl der Pro und Contra Argumente: "+count+'\n'+'\n')
```
