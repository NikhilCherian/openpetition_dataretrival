```python
def get_all_petitions(url):
    links=[]
    t=get_comment_linkpages(url)
    if len(t)==1:
        req1 = get(t[0])
        html_soup1 = BeautifulSoup(req1.text, 'html.parser')
        for l in html_soup1.find_all('li', class_ = 'clearfix'):
            if l.h3 != None:
                #print(l.h3.find('a',class_ ='theme-link-variant-3'))
                x=l.h3.find('a',class_ ='theme-link-variant-3')
                links.append(x['href'])
    else:
        for l in t:
    
            req1 = get(l)
            html_soup1 = BeautifulSoup(req1.text, 'html.parser')
            for l in html_soup1.find_all('li', class_ = 'clearfix'):
                if l.h3 != None:
                    #print(l.h3.find('a',class_ ='theme-link-variant-3'))
                    x=l.h3.find('a',class_ ='theme-link-variant-3')
                    links.append(x['href'])
    return links
```

```python
def get_data(url):
    #url_comment_syrien='https://www.openpetition.de/petition/kommentare/keine-abschiebungen-nach-syrien?sortierung=datum'
    import regex as re
    import lxml
    import requests
    #print("url",url)
    requests.packages.urllib3.disable_warnings()
    req_comment_syrien = requests.get(url)
    
    html_soup_comment_syrien = BeautifulSoup(req_comment_syrien.text, "lxml")
    type(html_soup_comment_syrien)
    
    sy_content=html_soup_comment_syrien.find_all("article",{'class':'kommentar clearfix'})
    #print(sy_content)
    
    name_date=[]
    for p in sy_content:
        #print(p.find("p",{'class':'date theme-text-variant-6'}).get_text(strip=True))
        n=p.find("p",{'class':'date theme-text-variant-6'}).get_text(strip=True)
        #n=slugify(n)#to remove the turkish alphabets
        name_date.append(n)
        
        
    gewicht=[]
    for p in sy_content:
        #print(p.find("span",{'class':'gewicht'}).text)
        gewicht.append(p.find("span",{'class':'gewicht'}).text)
        
    comments=[]
    for p in sy_content:
        #print(p.find("p",{'class':'theme-text-variant-5'}).get_text(strip=True))
        x=p.find("p",{'class':'theme-text-variant-5'}).get_text(strip=True)
        x=give_emoji_free_text(x)
        #print(x)
        #x=x.strip('[\U0001f54a][\][\U0001f54A][\']')
        comments.append(x)
    n_comments=len(comments)    
    for i in range(len(name_date)):
        #print(len(name_date[i].split()))
        #print(name_date[i].split())
        time=name_date[i].split()[-2]
        date=name_date[i].split()[-3]
        
        place=name_date[i].split()[-5]
        
        email = re.compile('(\w+)')
        #print("email",email.findall(place))
        #place=email.findall(place)[0]
        print(place)
        if email.findall(place):
            place=email.findall(place)[0]
        else:
            place="Nicht angegeben"
            
        name=name_date[i].split()[0:-5]
        
        comment=comments[i]
        name=listToString(name)
        
        gewichts=gewicht[i] 
        with open("openpetitions_erfolg.txt","a",errors='replace') as f: 
        
            f.write('\n'+'--------------------------------------------------')
            f.write('\n'+"Datum: "+date +'\n')
            f.write('Uhr: '+time+'\n')
            f.write('Name: '+name+'\n')
            f.write('Ort: '+place+'\n')
            f.write('Gewicht: '+gewichts+'\n')
            f.write('Kommentar: \n'+comment+'\n')
        
    #f.writelines(pet_description+'\n')

    f.close()
    return n_comments
```

```python
import emoji
def give_emoji_free_text(text):
    allchars = [str for str in text]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.split() if not any(i in str for i in emoji_list)])
    return clean_text
```

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    #s=s.replace('','')
    s=str1.join(s).replace('','')
    return (s)
```
